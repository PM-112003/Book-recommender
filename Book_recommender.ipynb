```python
book_ratings_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
book_ratings_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
book_ratings_df.sample(10)
```

```python
# Convert 'Book-Rating' to string to identify non-numeric values
ratings_with_name['Book-Rating'] = ratings_with_name['Book-Rating'].astype(str)

# Remove rows with non-numeric values
ratings_with_name = ratings_with_name[ratings_with_name['Book-Rating'].str.isnumeric()]

# Convert 'Book-Rating' to numeric
ratings_with_name['Book-Rating'] = pd.to_numeric(ratings_with_name['Book-Rating'], errors='coerce')

# Drop rows with NaN values in 'Book-Rating'
ratings_with_name.dropna(subset=['Book-Rating'], inplace=True)

# Convert 'Book-Rating' to integer
ratings_with_name['Book-Rating'] = ratings_with_name['Book-Rating'].astype(int)
```

## Collaborative filtering 

```python
ratings_with_name.groupby('User-ID').count()
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
knowledgeable_users = x[x].index
```

```python
knowledgeable_users
```

```python
# these are the ratings that are being done by the knowledgeable_users 
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(knowledgeable_users)]
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
#final ratings of those books whose ratings are above 50
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
final_ratings
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

The format of data we need for our filtering

```python
final_pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
final_pt.fillna(0,inplace=True)
final_pt.head(5)
```

Here every row is a point in an 810 dimensional plane

```python
# to find the similiarity between different data points we do the following
from sklearn.metrics.pairwise import cosine_similarity
cosine_similarity(final_pt)
```

```python
cosine_similarity(final_pt).shape
```

euclidean distance of 706 books with 706 books in the multi-dimensional plane

```python
similarity_scores = cosine_similarity(final_pt)
```

```python
def recommend(book_name):
    #index fetch
    index = np.where(final_pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]

    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == final_pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
```

```python
recommend('1984')
```
